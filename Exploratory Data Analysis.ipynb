```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Utility to normalize column names to predictable names
def normalize_cols(df):
    df = df.copy()
    df.columns = [c.strip().lower().replace(' ', '_').replace('-', '_') for c in df.columns]
    return df

DATA_PATH = Path(r'C:\Users\HP\Desktop\PYTHON PROJECT FROM DATALABS\appendicitis_clean.csv')
OUT_DIR = Path(r'C:\Users\HP\Desktop\PYTHON PROJECT FROM DATALABS\appendicitis_plots')
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Load data
try:
    df = pd.read_csv(DATA_PATH)
except Exception as e:
    raise FileNotFoundError(f"Could not open {DATA_PATH}. Error: {e}")

df = normalize_cols(df)

# Display top rows and info to the user using the helper from the environment
try:
    from caas_jupyter_tools import display_dataframe_to_user
    display_dataframe_to_user("Appendicitis dataset preview", df.head(50))
except Exception:
    # fallback: print head
    print("First 10 rows of the dataset:")
    display(df.head(10))

# Standardize a small set of expected column name variants (mapping common variants)
col_map = {
    'age': None,
    'gender': None,
    'weight': None,
    'bmi': None,
    'psoas_sign': None,
    'appendix_on_us': None,
    'alvarado_score': None,
    'pediatric_appendicitis_score': None,
    'diagnosis': None,
    'management': None,
    'severity': None
}

# attempt to find best matching columns in dataframe for each expected column
for expected in list(col_map.keys()):
    # exact match
    if expected in df.columns:
        col_map[expected] = expected
        continue
    # try common variants
    variants = [
        expected,
        expected.replace('_', ''),
        expected.replace('_', '' ) + '_score',
        expected + '_score',
        expected + 's',
        expected.title(),
        expected.upper()
    ]
    found = None
    for v in variants:
        if v.lower() in df.columns:
            found = v.lower()
            break
    if found:
        col_map[expected] = found
    else:
        # try fuzzy contains
        for c in df.columns:
            if expected in c:
                col_map[expected] = c
                break

# short helper
def has(col):
    return col_map.get(col) in df.columns if col_map.get(col) else False

# Show detected column mapping
print("Detected column mapping (expected -> actual):")
for k,v in col_map.items():
    print(f"  {k:30s} -> {v}")

# Convert numeric-looking columns to numeric where appropriate
numeric_candidates = ['age','weight','bmi','alvarado_score','pediatric_appendicitis_score']
for nc in numeric_candidates:
    if col_map.get(nc) and col_map[nc] in df.columns:
        df[col_map[nc]] = pd.to_numeric(df[col_map[nc]], errors='coerce')

# Quick summary
print("\nDataset shape:", df.shape)
print("\nMissing values per selected column:")
for k,v in col_map.items():
    if v and v in df.columns:
        print(f"  {v}: {df[v].isna().sum()} missing")

saved_files = []

# 1) Age distribution histogram
if has('age'):
    plt.figure(figsize=(8,5))
    df[col_map['age']].dropna().hist(bins=20)
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.title('Age distribution')
    f = OUT_DIR / 'age_histogram.png'
    plt.tight_layout()
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 2) Gender counts bar chart
if has('gender'):
    plt.figure(figsize=(6,4))
    df[col_map['gender']].fillna('Unknown').value_counts().plot(kind='bar')
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.title('Gender counts')
    plt.tight_layout()
    f = OUT_DIR / 'gender_counts.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 3) Weight distribution histogram
if has('weight'):
    plt.figure(figsize=(8,5))
    df[col_map['weight']].dropna().hist(bins=20)
    plt.xlabel('Weight')
    plt.ylabel('Count')
    plt.title('Weight distribution')
    plt.tight_layout()
    f = OUT_DIR / 'weight_histogram.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 4) BMI distribution histogram
if has('bmi'):
    plt.figure(figsize=(8,5))
    df[col_map['bmi']].dropna().hist(bins=20)
    plt.xlabel('BMI')
    plt.ylabel('Count')
    plt.title('BMI distribution')
    plt.tight_layout()
    f = OUT_DIR / 'bmi_histogram.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 5) psoas_sign counts (assume categorical boolean-like)
if has('psoas_sign'):
    plt.figure(figsize=(6,4))
    df[col_map['psoas_sign']].fillna('Unknown').value_counts().plot(kind='bar')
    plt.xlabel('Psoas sign')
    plt.ylabel('Count')
    plt.title('Psoas sign counts')
    plt.tight_layout()
    f = OUT_DIR / 'psoas_sign_counts.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 6) appendix_on_us counts
if has('appendix_on_us'):
    plt.figure(figsize=(6,4))
    df[col_map['appendix_on_us']].fillna('Unknown').value_counts().plot(kind='bar')
    plt.xlabel('Appendix on ultrasound (appendix_on_us)')
    plt.ylabel('Count')
    plt.title('Appendix on US - counts')
    plt.tight_layout()
    f = OUT_DIR / 'appendix_on_us_counts.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 7) Alvarado score distribution
if has('alvarado_score'):
    plt.figure(figsize=(8,5))
    df[col_map['alvarado_score']].dropna().hist(bins=range(0,13))
    plt.xlabel('Alvarado score')
    plt.ylabel('Count')
    plt.title('Alvarado score distribution')
    plt.tight_layout()
    f = OUT_DIR / 'alvarado_score_histogram.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 8) Pediatric appendicitis score distribution
if has('pediatric_appendicitis_score'):
    plt.figure(figsize=(8,5))
    df[col_map['pediatric_appendicitis_score']].dropna().hist(bins=20)
    plt.xlabel('Pediatric appendicitis score')
    plt.ylabel('Count')
    plt.title('Pediatric appendicitis score distribution')
    plt.tight_layout()
    f = OUT_DIR / 'pediatric_appendicitis_score_histogram.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 9) Diagnosis counts
if has('diagnosis'):
    plt.figure(figsize=(8,5))
    df[col_map['diagnosis']].fillna('Unknown').value_counts().plot(kind='bar')
    plt.xlabel('Diagnosis')
    plt.ylabel('Count')
    plt.title('Diagnosis counts')
    plt.tight_layout()
    f = OUT_DIR / 'diagnosis_counts.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 10) Management counts
if has('management'):
    plt.figure(figsize=(8,5))
    df[col_map['management']].fillna('Unknown').value_counts().plot(kind='bar')
    plt.xlabel('Management')
    plt.ylabel('Count')
    plt.title('Management counts')
    plt.tight_layout()
    f = OUT_DIR / 'management_counts.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# 11) Severity counts
if has('severity'):
    plt.figure(figsize=(8,5))
    df[col_map['severity']].fillna('Unknown').value_counts().plot(kind='bar')
    plt.xlabel('Severity')
    plt.ylabel('Count')
    plt.title('Severity counts')
    plt.tight_layout()
    f = OUT_DIR / 'severity_counts.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# Bivariate: boxplots of Age by Diagnosis and Severity
if has('age') and has('diagnosis'):
    plt.figure(figsize=(10,6))
    # prepare data dropping NA diagnosis
    subset = df[[col_map['age'], col_map['diagnosis']]].dropna()
    groups = subset.groupby(col_map['diagnosis'])[col_map['age']].apply(list)
    plt.boxplot(groups.tolist(), labels=groups.index, vert=True)
    plt.ylabel('Age')
    plt.xlabel('Diagnosis')
    plt.title('Age by Diagnosis (boxplot)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    f = OUT_DIR / 'age_by_diagnosis_boxplot.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

if has('bmi') and has('severity'):
    plt.figure(figsize=(10,6))
    subset = df[[col_map['bmi'], col_map['severity']]].dropna()
    groups = subset.groupby(col_map['severity'])[col_map['bmi']].apply(list)
    plt.boxplot(groups.tolist(), labels=groups.index, vert=True)
    plt.ylabel('BMI')
    plt.xlabel('Severity')
    plt.title('BMI by Severity (boxplot)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    f = OUT_DIR / 'bmi_by_severity_boxplot.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# Scatter: Weight vs BMI colored by Diagnosis (use marker and default color cycle)
if has('weight') and has('bmi') and has('diagnosis'):
    plt.figure(figsize=(8,6))
    subset = df[[col_map['weight'], col_map['bmi'], col_map['diagnosis']]].dropna()
    # plot each diagnosis group separately so legend is meaningful
    diagnoses = subset[col_map['diagnosis']].unique()
    for diag in diagnoses:
        g = subset[subset[col_map['diagnosis']] == diag]
        plt.scatter(g[col_map['weight']], g[col_map['bmi']], label=str(diag), alpha=0.7)
    plt.xlabel('Weight')
    plt.ylabel('BMI')
    plt.title('Weight vs BMI by Diagnosis')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
    plt.tight_layout()
    f = OUT_DIR / 'weight_vs_bmi_by_diagnosis.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# Correlation matrix heatmap for numeric columns (Age, Weight, BMI, Alvarado score, Pediatric score)
numeric_cols = []
for name in ['age','weight','bmi','alvarado_score','pediatric_appendicitis_score']:
    if has(name):
        numeric_cols.append(col_map[name])
if numeric_cols:
    corr = df[numeric_cols].corr()
    plt.figure(figsize=(6,5))
    plt.imshow(corr, interpolation='nearest', aspect='auto')
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha='right')
    plt.yticks(range(len(corr.index)), corr.index)
    plt.title('Correlation matrix (numeric columns)')
    plt.tight_layout()
    f = OUT_DIR / 'correlation_matrix.png'
    plt.savefig(f)
    plt.show()
    saved_files.append(str(f))

# List saved files and give simple summary stats
print("\nSaved plot files:")
for sf in saved_files:
    print(" ", sf)

print("\nSummary statistics for key numeric columns:")
display(df[numeric_cols].describe())

# If no saved files were produced, warn the user
if not saved_files:
    print("\nNo plots were generated because none of the expected columns were found. Please check column names.")
else:
    print("\nPlots saved in:", OUT_DIR)
    # show small sample of the cleaned dataframe columns for user's review
    print("\nColumns available in the cleaned dataframe:")
    print(df.columns.tolist())

# End of script. The produced PNGs are in /mnt/data/appendicitis_plots/ and can be downloaded from the notebook UI.

```
